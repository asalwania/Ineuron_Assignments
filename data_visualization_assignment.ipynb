```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris=sns.load_dataset('iris')
print(iris.shape)
iris.head()
```

```python
pca=PCA(3)
pc_data=pca.fit_transform(iris.drop('species',axis=1))
```

```python
pc_iris=pd.DataFrame(pc_data,columns=['PC1','PC2','PC3'])
pc_iris["species"]=iris.species
pc_iris.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('PC1', fontsize = 15)
ax.set_ylabel('PC2', fontsize = 15)
ax.set_zlabel('PC3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

targets = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = pc_iris['species'] == target
    ax.scatter(pc_iris.loc[indicesToKeep, 'PC1']
               , pc_iris.loc[indicesToKeep, 'PC2']
               , pc_iris.loc[indicesToKeep, 'PC3']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.show()
```
